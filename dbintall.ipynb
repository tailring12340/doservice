```python
import pymysql

# 데이터베이스 연결 설정
host = 'localhost'
user = 'root'
password = '0000'
database = 'japonica'

# 데이터베이스 연결
connection = pymysql.connect(host=host, user=user, password=password, database=database)

# 연결 성공 시 추가 작업 수행
if connection:
    print("Mariadb에 성공적으로 연결되었습니다.")
```

```python
import pandas as pd
japonica_training_food_supply_tb = pd.read_csv("E:/kimwoojae/3학기/doservice/dataset/japonica_training_food_supply_tb.csv")
japonica_training_sensor_val_tb = pd.read_csv("E:/kimwoojae/3학기/doservice/dataset/japonica_training_sensor_val_tb.csv")
```

```python
japonica_training_sensor_val_tb.info()
```

```python
japonica_training_sensor_val_tb['Unnamed: 0']
```

```python
cursor = connection.cursor()

for _, row in japonica_training_sensor_val_tb.iterrows():
    insert_query = f'''
    INSERT INTO sensor (farm_id, tank_id, mea_dt, do_mg, do_temp, ph, orp, co2_mg, air_oxy, light_ma)
    VALUES ({row['farm_id']}, '{row['tank_id']}', '{row['mea_dt']}', '{row['do_mg']}', '{row['do_temp']}', '{row['ph']}', '{row['orp']}', '{row['co2_mg']}', '{row['air_oxy']}', '{row['light_ma']}')
    '''
    cursor.execute(insert_query)

connection.commit()

cursor.close()
print("데이터가 성공적으로 데이터베이스에 저장되었습니다.")
```

```python
japonica_training_food_supply_tb['feed_dt'] = pd.to_datetime(japonica_training_food_supply_tb['feed_dt'], format='%Y%m%d%H%M', errors='raise')
japonica_training_food_supply_tb["feed_dt"] = japonica_training_food_supply_tb["feed_dt"].apply(str)
japonica_training_food_supply_tb["feed_dt"] = japonica_training_food_supply_tb["feed_dt"].str.slice(start=0, stop=16)
```

```python
japonica_training_food_supply_tb
```

```python
japonica_training_food_supply_tb = japonica_training_food_supply_tb.fillna(0)
```

```python
cursor = connection.cursor()

for _, row in japonica_training_food_supply_tb.iterrows():
    insert_query = f'''
    INSERT INTO food_supply (farm_id, tank_id, feed_dt, feed_quantity, water_quantity)
    VALUES ({row['farm_id']}, '{row['tank_id']}', '{row['feed_dt']}', '{row['feed_quantity']}', '{row['water_quantity']}')
    '''
    cursor.execute(insert_query)

connection.commit()

cursor.close()
print("데이터가 성공적으로 데이터베이스에 저장되었습니다.")
```
